```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
import tensorflow as tf
```

```python
normal = glob.glob('../dataset/train/normal/*.jpeg')
pneumonia = glob.glob('../dataset/train/pneumonia/*.jpeg')
```

```python
plt.figure(figsize=(8, 8))
labels = "normal"
for i, images in enumerate(normal[:9]):
    ax = plt.subplot(3, 3, i + 1)
    img = cv2.imread(images)
    plt.imshow(img)
    plt.title(f'{labels} {img.shape}')
    plt.axis("off")

plt.figure(figsize=(8, 8))
labels = "pneumonia"
for i, images in enumerate(pneumonia[:9]):
    ax = plt.subplot(3, 3, i + 1)
    img = cv2.imread(images)
    plt.imshow(img)
    plt.title(f'{labels} {img.shape}')
    plt.axis("off")
```

# <font  color = pink> 이미지 전처리

```python
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

training_set = train_datagen.flow_from_directory('../dataset/train',target_size=(64, 64), batch_size=32, class_mode='binary')
val_set = test_datagen.flow_from_directory('../dataset/val', target_size=(64, 64),batch_size=32, class_mode='binary')
test_set = test_datagen.flow_from_directory('../dataset/test', target_size=(64, 64),batch_size=32, class_mode='binary')
```

# <font color = pink> CNN 모델 구성

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
mc = ModelCheckpoint('pneumonia_best.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(training_set, epochs=100, callbacks=[es, mc], batch_size=32, validation_data = val_set)
```

```python
from tensorflow.keras.models import load_model

model = load_model('pneumonia_model.h5')

# 테스트 데이터셋으로 모델을 평가합니다.
test_loss, test_accuracy = model.evaluate(test_set)

# 정확도를 출력합니다.
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```

```python
model.save('pneumonia_model.h5')
```

# <font color = pink> DenseNet201 모델 테스트

```python
def create_gen():
    train_generator = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input,
        validation_split=0.1
    )

    test_generator = tf.keras.preprocessing.image.ImageDataGenerator(
        preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input
    )
    
    return train_generator,test_generator
```

# <font color = pink> 모델 테스트

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# 훈련된 모델 불러오기
test_normal = glob.glob('../dataset/test/normal/*.jpeg')
test_pneumonia = glob.glob('../dataset/test/pneumonia/*.jpeg')

model = load_model('pneumonia_model.h5')

# 예측할 이미지 불러오기 및 전처리
image_path = test_pneumonia[0]
img = image.load_img(image_path, target_size=(64, 64))
img = image.img_to_array(img)
img = np.expand_dims(img, axis=0)
img = img / 255.0  # 이미지를 0에서 1 사이의 값으로 정규화


prediction = model.predict(img)

if prediction[0][0] >= 0.5: # 0.5 이상이라면 폐렴, 미만이면 정상
    print("이 이미지는 폐렴(pneumonia)일 가능성이 있습니다.")
else:
    print("이 이미지는 정상(normal)일 가능성이 있습니다.")
```

```python
# 5. 모델 학습과정 표시하기
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.plot(history.history['loss'], 'y', label = 'train loss')
loss_ax.plot(history.history['val_loss'], 'r', label = 'val loss')

acc_ax.plot(history.history['accuracy'], 'b', label = 'train accuracy')
acc_ax.plot(history.history['val_accuracy'], 'g', label = 'val accuracy')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc = 'upper left')
acc_ax.legend(loc = 'lower left')

plt.show()
```
